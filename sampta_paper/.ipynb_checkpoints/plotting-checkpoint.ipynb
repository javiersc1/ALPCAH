# Libraries

```julia
using Plots
using Random
using Distributions
using LinearAlgebra
include("pogm_restart.jl")
using ProgressMeter
using HePPCAT
using Distributed
using JLD
```

# Known Variance Data Import

```julia
varianceRatio = range( 1, 100,20) #20 for plot
pointRatio = range(1,100,20); # 20 points for plot
```

```julia
tickVector = vcat(1:9:100);
```

```julia
knownVarianceHPCA = load("knownVarianceMatrix.jld", "hpcaMatrix");
```

```julia
knownVariancePPCA = load("knownVarianceMatrix.jld", "ppcaMatrix");
```

```julia
knownVariancePPCA_GOOD = load("knownVarianceMatrix.jld", "goodsubspaceMatrix");
```

# KNOWN VARIANCE PLOTS

```julia
heatmap(pointRatio, varianceRatio, knownVarianceHPCA./knownVariancePPCA, title="HPCA/PCA [KNOWN VAR]", xlabel="Point Ratio N2/N1, N1=10", ylabel="Variance Ratio ν2/ν1, ν1=1",yflip=true, xticks=tickVector, yticks = tickVector)
#savefig("HPCA_KNOWN_PCA_HEATMAP.pdf")
```

```julia
heatmap(pointRatio, varianceRatio, knownVarianceHPCA./knownVariancePPCA_GOOD, title="HPCA/PCA-GOOD [KNOWN VAR]", xlabel="Point Ratio N2/N1, N1=10", ylabel="Variance Ratio ν2/ν1, ν1=1",yflip=true,xticks=tickVector, yticks = tickVector)
#savefig("HPCA_KNOWN_PCA_GOOD_HEATMAP.pdf")
```

```julia
heatmap(pointRatio, varianceRatio, knownVarianceHPCA, title="HPCA [KNOWN VAR]", xlabel="Point Ratio N2/N1, N1=10", ylabel="Variance Ratio ν2/ν1, ν1=1",yflip=true, xticks=tickVector, yticks = tickVector)
#savefig("UNKNOWN_HEPPCAT.pdf")
```
