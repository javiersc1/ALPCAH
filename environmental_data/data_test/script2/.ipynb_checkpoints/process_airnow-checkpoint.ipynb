```julia
using NPZ
airnow = npzread("airMatrix.npy")
```

```julia
using Plots
using Impute
function nantomissing(value)
    if isnan(value)==true
        return missing
    else
        return value
    end
end
airnow = nantomissing.(airnow)
```

```julia
airnow2 = Impute.interp(airnow; dims=2) |> Impute.locf(;dims=2) |> Impute.nocb(;dims=2)
```

```julia
heatmap(airnow2)
```

```julia
plot(mean(airnow2, dims=2), label="airnow mean")
plot!(title="California Data - Mine using Kyle's method", xlabel="Time (~1week)", ylabel="PM2.5 Concenctration")
```

```julia
using JLD
save("airMatrix.jld","airMatrix", airnow2)
```
