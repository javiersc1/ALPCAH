```julia
using Plots
using LinearAlgebra
using Random
using JLD
using ProgressMeter
using Statistics
using MIRT
include("../alpcahFamily.jl")
include("../helperFunctions.jl")
```

```julia
function POGM_RPCA(Y, λr, r; niter::Int=100)
    M,N = size(Y)
    A = hcat(Matrix(1.0I, M, N),Matrix(1.0I, M, N))
    B = vcat(A,A)
    Lf = opnorm(B)
    x0 = vcat(zeros(size(Y)),zeros(size(Y)))
    grad = K -> repeat(-1*(Y-K[1:M,:]-K[M+1:end,:]),2,1)
    soft = (x,t) -> sign.(x) .* max.(abs.(x) .- t, 0)
    function svst(x,t)
        U,S,Vt = svd(x)
        return U*diagm(soft.(S, t))*Vt'
    end
    prox1 = (z,c) -> vcat(svst(z[1:M,:], c*λr), soft(z[M+1:end,:], c))
    #prox1 = (z,c) -> vcat(svst(z[1:M,:], c), soft(z[M+1:end,:], λr*c))
    K, _ = pogm_restart(x0, x->0, grad, Lf ; g_prox=prox1, niter=niter, mom=:fpgm, restart=:gr)
    K = K[1:M,:] #,K[M+1:end,:]  Outlier matrix
    return svd(K).U[:,1:r]
end
```

```julia
r = 3
U = generateSubspace(100, r; seed=1234)
v = [0.01; 100]
p = [10; 100]
Y = generateData(U, v, p; coordinateWindow= 10, coordinateType= :uniform, seed=4321);
```

```julia
U_ALPCAH = LR_ALPCAH(Y,r; alpcahIter= 100)
affinityError(U, U_ALPCAH)
```

```julia
U_PCA = PCA(Y,r)
affinityError(U, U_PCA)
```

```julia
U_RPCA = POGM_RPCA(Y, 20, r; niter=100)
affinityError(U,U_RPCA)
```

```julia
function validateRPCA(Ytrain, Utrain, r, niter)
    λ = [0.001; 0.01; 0.1; 1; 10; 15; 20; 25; 30; 40; 50; 60; 70; 80; 90; 100]
    errors = zeros(length(λ))
    Threads.@threads for k=1:length(λ)
        U_RPCA = POGM_RPCA(Ytrain, λ[k], r; niter=niter)
        errors[k] = affinityError(Utrain, U_RPCA)
    end
    λ1 = λ[argmin(errors)]
    return λ1
end
```

```julia
v1 = 0.1
vr = range(1, 1000, 25)
trials = 10
methods = 3
r = 3
d = 100
p = [35; 65] # 10,100
niter = 100
```

```julia
stateToggle = 0; λ1=1;
progress = Progress(length(vr))
dataMatrix = zeros(trials, methods, length(vr));
for j=1:length(vr)
   for i = 1:trials
        # Data Generate
        Utest = generateSubspace(d, r)
        Ytest = generateData(Utest, [v1; vr[j]*v1], p);
        if (stateToggle == 0)
            Utrain = generateSubspace(d, r)
            Ytrain = generateData(Utrain, [v1; vr[j]*v1], p)
            λ1 = validateRPCA(Ytrain, Utrain, r, niter)
            println(λ1)
            stateToggle = 1
        end
        # PCA
        U_PCA = PCA(Ytest, r)
        dataMatrix[i,1,j] = affinityError(Utest, U_PCA)
        # LR-ALPCAH
        U_ALPCAH = LR_ALPCAH(Ytest, r; alpcahIter=niter)
        dataMatrix[i,2,j] = affinityError(Utest, U_ALPCAH)
        # RPCA
        U_RPCA = POGM_RPCA(Ytest, λ1, r; niter=niter)
        dataMatrix[i,3,j] = affinityError(Utest, U_RPCA)
    end
    stateToggle = 0
    next!(progress)
end
finish!(progress)
```

```julia
meanMatrix = zeros(methods, length(vr))
stdMatrix = zeros(methods, length(vr))
for i = 1:methods
    meanMatrix[i,:] = mean(dataMatrix[:,i,:], dims=1)
    stdMatrix[i,:] = stdm(dataMatrix[:,i,:], meanMatrix[i,:]'; dims=1, corrected=false)
end
```

```julia
plot(title="Outliers vs Heteroscedasticity", xlabel="ν2/ν1, ν1=$v1", ylabel="Subspace Affinity Error")
plot!(vr, meanMatrix[1,:], ribbon=1*stdMatrix[1,:], label="PCA", linewidth=2, fillalpha=0.2)
plot!(vr, meanMatrix[2,:], ribbon=1*stdMatrix[2,:], label="ALPCAH", linewidth=2, fillalpha=0.2)
plot!(vr, meanMatrix[3,:], ribbon=1*stdMatrix[3,:], label="RPCA", linewidth=2, fillalpha=0.2)
#savefig("rpca_variance_graph.pdf")
```
