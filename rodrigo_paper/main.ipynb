# Smooth ALPCAH

```julia
function hyperbolaGrad(X, δ, rank)
    T = svd(X)
    S = copy(T.S)
    S[(rank+1):end] .= gauge_derivative_hyperbola.(S[(rank+1):end],δ)
    S[1:rank] .= 1.0
    output = T.U*Diagonal(S)*T.Vt
    return output;
end

function gauge_derivative_hyperbola(S,δ)
    output = S / (δ * sqrt(1 + (S / δ)^2))
    return output
end

hyperbola_function = (S,δ) -> δ^1 * sqrt.(1 .+ (S ./ δ).^2)
```

```julia
function ALPCAH_UNKNOWN_APGD(Y::Matrix, rank::Int, λr::Real; alpcahIter::Int = 200, apgdIter::Int = 5, varianceMethod::Symbol = :groupless, varfloor::Real=1e-9)
    D, N = size(Y)
    v = zeros(N)
    # Krylov-based Lanczos Algorithm
    U = PCA(Y,rank) #tsvd(Y, rank)[1]
    X = U*U'*Y
    #variance method initialization
    if varianceMethod === :groupless
        v = grouplessVarianceUpdate(Y, X; varfloor=varfloor)
    else
        #v = groupedVarianceUpdate(Y, L*R' ; varfloor=varfloor)
    end
    Π = v.^-1
    Lf = maximum(Π)
    Π = Diagonal(Π)
    X = zeros(size(Y)) # weirdly necessary to get lower error
    grad = K -> -1*(Y-K)*Π
    #alpha = LinRange(0.0, λr, alpcahIter)
    for i=1:alpcahIter
        # apgd
        g_prox = (z,c) -> TSVT(z, c*λr, rank) #TSVT(z, c*λr, rank) # TSVT(z, c*alpha[i], rank)
        X, _ = pogm_restart(X, x -> 0, grad, Lf ; mom=:fpgm, g_prox, niter=apgdIter);
        # variance updates
        if varianceMethod === :groupless
            v = grouplessVarianceUpdate(Y, X; varfloor=varfloor)
        else
            #v = groupedVarianceUpdate(Y, L*R' ; varfloor=varfloor)
        end
        Π = v.^-1
        Lf = maximum(Π)
        Π = Diagonal(Π)
    end
    # extract left vectors from X
    U = PCA(X, rank) #tsvd(X, rank)[1]
    return U
end

function ALPCAH_UNKNOWN_OGM(Y::Matrix, rank::Int, λr::Real, δ::Real; alpcahIter::Int = 200, apgdIter::Int = 5, varianceMethod::Symbol = :groupless, varfloor::Real=1e-9)
    D, N = size(Y)
    v = zeros(N)
    # Krylov-based Lanczos Algorithm
    U = PCA(Y, rank) #tsvd(Y, rank)[1]
    X = U*U'*Y
    #variance method initialization
    if varianceMethod === :groupless
        v = grouplessVarianceUpdate(Y, X; varfloor=varfloor)
    else
        #v = groupedVarianceUpdate(Y, L*R' ; varfloor=varfloor)
    end
    Π = v.^-1
    ϕ_L = 1/δ
    Lf = maximum(Π) + λr*ϕ_L;
    Π = Diagonal(Π)
    X = zeros(size(Y)) # weirdly necessary to get lower error
    grad = K -> -1*(Y-K)*Π + λr*hyperbolaGrad(K, δ, rank);
    #alpha = LinRange(0.0, λr, alpcahIter)
    for i=1:alpcahIter
        # apgd
        #g_prox = (z,c) -> TSVT(z, c*λr, rank) # TSVT(z, c*alpha[i], rank)
        X, _ = pogm_restart(X, (x) -> 0, grad, Lf ; mom=:pogm, niter=apgdIter, f_mu=0.0);
        # variance updates
        if varianceMethod === :groupless
            v = grouplessVarianceUpdate(Y, X; varfloor=varfloor)
        else
            #v = groupedVarianceUpdate(Y, L*R' ; varfloor=varfloor)
        end
        Π = v.^-1
        Lf = maximum(Π) + λr*ϕ_L;
        Π = Diagonal(Π)
    end
    # extract left vectors from X
    U = PCA(X, rank) #tsvd(X, rank)[1]
    return U
end

function grouplessVarianceUpdate(Y::Matrix, X::Matrix; varfloor::Real=1e-9)
    D= size(Y)[1]
    Π = (1/D)*norm.(eachcol(Y - X)).^2
    return max.(Π, varfloor)
end

function TSVT(A::Matrix,λ::Real,k::Int)
    U,S,V = svd(A)
    S[(k+1):end] = softTresh.(S[(k+1):end],λ)
    return U*Diagonal(S)*V'
end

function softTresh(x::Real, λ::Real)
    return sign(x) * max(abs(x) - λ, 0)
end
```

# Experiment

```julia
trials = 10;
D = 100;
d = 10;
v = [0.25, 100];
N = [10, 490];
U = generateSubspace(D, d; seed=0)
Y = generateData(U, v, N; coordinateWindow=33, seed=1);
varianceVector = vcat(v[1]*ones(N[1]), v[2]*ones(N[2]));
YL = []
for i = 1:trials
    Y0 = generateData(U, v, N; coordinateWindow=33, seed=i+1)
    push!(YL,Y0)
end
```

```julia
wpca_list = []; pca_list = []; alpcah_nonsmooth_list = []; alpcah_smooth1_list = []; alpcah_smooth2_list = []; alpcah_smooth3_list = [];
#λr = vcat(0.0:5:100.0); λr[1] = 0.1;
λr = vcat(0.0:2.0:20); λr[1] = 0.01;
for i=1:trials
    println(i)
    flush(stdout)
    # PCA
    U_PCA = PCA(YL[i], d)
    errorPCA = affinityError(U, U_PCA)
    push!(pca_list, errorPCA)
    # Weighted PCA [known variance]
    U_WPCA = weightedPCA(YL[i], varianceVector, d) 
    errorWPCA = affinityError(U, U_WPCA)
    push!(wpca_list, errorWPCA)
    # ALPCAH [unknown variance - APGD]
    errorALPCAH = zeros(size(λr))
    Threads.@threads for j=1:length(λr)
        U_ALPCAH = ALPCAH_UNKNOWN_APGD(YL[i], d, λr[j]; alpcahIter= 200,apgdIter= 5) 
        errorALPCAH[j] = affinityError(U, U_ALPCAH)
    end
    push!(alpcah_nonsmooth_list, errorALPCAH)
    # smooth 1
    errorALPCAH = zeros(size(λr))
    Threads.@threads for j=1:length(λr)
        U_ALPCAH = ALPCAH_UNKNOWN_OGM(YL[i],d, 0.1*λr[j], 1; alpcahIter= 200,apgdIter= 5) 
        errorALPCAH[j] = affinityError(U, U_ALPCAH)
    end
    push!(alpcah_smooth1_list, errorALPCAH)
     # smooth 2
    errorALPCAH = zeros(size(λr))
    Threads.@threads for j=1:length(λr)
        U_ALPCAH = ALPCAH_UNKNOWN_OGM(YL[i],d, 0.1*λr[j], 0.1; alpcahIter= 200,apgdIter= 5) 
        errorALPCAH[j] = affinityError(U, U_ALPCAH)
    end
    push!(alpcah_smooth2_list, errorALPCAH)
     # smooth 3
    errorALPCAH = zeros(size(λr))
    Threads.@threads for j=1:length(λr)
        U_ALPCAH = ALPCAH_UNKNOWN_OGM(YL[i],d,0.1*λr[j], 10; alpcahIter= 200,apgdIter= 5) 
        errorALPCAH[j] = affinityError(U, U_ALPCAH)
    end
    push!(alpcah_smooth3_list, errorALPCAH)
end
```

```julia
using Plots
using LaTeXStrings
pca_mean = mean(pca_list); wpca_mean = mean(wpca_list); 
alpcah_nonsmooth_mean = mean(alpcah_nonsmooth_list); alpcah_smooth1_mean = mean(alpcah_smooth1_list);
alpcah_smooth2_mean = mean(alpcah_smooth2_list); alpcah_smooth3_mean = mean(alpcah_smooth3_list);
colors = theme_palette(:auto).colors
```

```julia
plot(xlabel="λ (regularization parameter)",ylim=(0.17,0.35),ylabel=L"\Vert U_{*} U_{*}^{'}- UU' \Vert_F/ \Vert UU' \Vert_F", title="Subspace Approximation - Low Rank Models")
plot!(legend_position=:topright, legendfontpointsize=10,xtickfontsize=12,ytickfontsize=12,guidefontsize=16,titlefontsize=14)
# NO λ METHODS
hline!([pca_mean],ribbon=(pca_mean-minimum(pca_list),maximum(pca_list)-pca_mean),c=colors[4], fillalpha = 0.0,linewidth=3, label="PCA")
hline!([wpca_mean],ribbon=(wpca_mean-minimum(wpca_list),maximum(wpca_list)-wpca_mean), c=:black, linewidth=3,fillalpha = 0.0, label="Weighted PCA")
# ALPCAH-nonsmooth
maxValue = vec(maximum(reduce(vcat,transpose.(alpcah_nonsmooth_list)),dims=1))
minValue = vec(minimum(reduce(vcat,transpose.(alpcah_nonsmooth_list)),dims=1))
plot!(λr,alpcah_nonsmooth_mean, ribbon=(alpcah_nonsmooth_mean .- minValue, maxValue .- alpcah_nonsmooth_mean), fillalpha = 0.0, c=colors[1], linewidth=3, label="ALPCAH-NUCLEAR", marker=:x)
# ALPCAH-smooth
maxValue = vec(maximum(reduce(vcat,transpose.(alpcah_smooth1_list)),dims=1))
minValue = vec(minimum(reduce(vcat,transpose.(alpcah_smooth1_list)),dims=1))
plot!(λr,alpcah_smooth1_mean, ribbon=(alpcah_smooth1_mean .- minValue, maxValue .- alpcah_smooth1_mean), fillalpha = 0.0, c=colors[2], linewidth=3, label="ALPCAH-HYPERBOLA (δ=1)", marker=:diamond)
# ALPCAH-smooth 2
maxValue = vec(maximum(reduce(vcat,transpose.(alpcah_smooth2_list)),dims=1))
minValue = vec(minimum(reduce(vcat,transpose.(alpcah_smooth2_list)),dims=1))
plot!(λr,alpcah_smooth2_mean, ribbon=(alpcah_smooth2_mean .- minValue, maxValue .- alpcah_smooth2_mean), fillalpha = 0.0, c=colors[3], linewidth=3, label="ALPCAH-HYPERBOLA (δ=0.1)", marker=:utriangle)
# ALPCAH-smooth 3
maxValue = vec(maximum(reduce(vcat,transpose.(alpcah_smooth3_list)),dims=1))
minValue = vec(minimum(reduce(vcat,transpose.(alpcah_smooth3_list)),dims=1))
plot!(λr,alpcah_smooth3_mean, ribbon=(alpcah_smooth3_mean .- minValue, maxValue .- alpcah_smooth3_mean), fillalpha = 0.0, c=colors[5], linewidth=3, label="ALPCAH-HYPERBOLA (δ=10)", marker=:cross)
#savefig("alpcah_comparisons.pdf")
```
