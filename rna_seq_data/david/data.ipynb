```julia
using MAT
using JLD
```

```julia
Y = matread("original_data.mat")["dropSeq"]
```

```julia
M, N = size(Y)
Ynew = Y .- ((1/M) .* sum(Y, dims=1))
```

```julia
weight = sqrt.(sum(Ynew.^2, dims=1))
Ynew = Ynew./(ones(M)*weight)
```

```julia
Ynew = [ isnan(x) ? 0 : x for x in Ynew ]
```

```julia
sum(isnan.(Ynew))
```

```julia
sum(isinf.(Ynew))
```

```julia
Ynew = Float64.(Ynew);
Ynew = 1e3 .* Ynew
```

```julia
using Random
index = randperm(5000)[1:5000];
```

```julia
Ytrain = Ynew[:,index[1:2500]]
Ytest = Ynew[:,index[2501:end]]
save("original_data.jld", "train", Ytrain, "test", Ytest)
```
