```julia
using LinearAlgebra
using HePPCAT
using Plots
using ProgressMeter
using Random, Statistics, StatsPlots
using Plots.PlotMeasures
using BenchmarkTools
include("../alpcahFamily.jl")
```

```julia
using PyCall
@pyimport pickle

function mypickle(filename, obj)
    out = open(filename,"w")
    pickle.dump(obj, out)
    close(out)
 end

function myunpickle(filename)
    r = nothing
    @pywith pybuiltin("open")(filename,"rb") as f begin
        r = pickle.load(f)
    end
    return r
end
```

```julia
data = myunpickle("airData.pickle")[2]

data[1] .= max.(data[1],0)
data[2] .= max.(data[2],0)
# data[2] .= min.(data[2],250) ### cap maximum reading
data[1] .-= mean(data[1],dims=2)
data[2] .-= mean(data[2],dims=2);
replace!(data[2], NaN=>0)

n1 = size(data[1])[2]
n2 = size(data[2])[2]

k = 30 ### the rank of the experiment
```

```julia
#Random.seed!(0)
ntrain1 = 25 ### number of training samples for AirNow
ntrain2 = 250 ### number of training samples for PurpleAir

data[1] .= data[1][:,shuffle(1:end)]
data[2] .= data[2][:,shuffle(1:end)]
train1 = data[1][:,1:ntrain1]
test1 = data[1][:,ntrain1+1:end]

train2 = data[2][:,1:ntrain2]
test2 = data[2][:,ntrain2+1:end]

Ytrain = [train1,train2]
Ytest = [test1,test2]
Ymat_train = hcat(Ytrain...);
Ymat_test = hcat(Ytest...);
    
ntest1 = size(Ytest[1])[2]
ntest2 = size(Ytest[2])[2]
ntest = ntest1 + ntest2

n = [size(Yl,2) for Yl in Ytrain]
d = size(Ytrain[1],1)
L = length(Ytrain)

niters = 100
Ylist_train = []
for t=1:(ntrain1+ntrain2)
    push!(Ylist_train, Ymat_train[:,t])
end
```

```julia
total = length([1,10,50]) #200 ###number of random training/test sets 

### Logs to record experiment stats
num_methods = 5
```

```julia
timeVector = zeros(num_methods,total)
test_errors = zeros(num_methods,total)
count = 1
for iter in [1,10,50]
    # PCA
    U_PCA = PCA(Ymat_train, k)
    timeVector[1,count] = @belapsed PCA(Ymat_train, k)

    # HEPPCAT
    Mheppcat = heppcat(Ylist_train,k,niters; varfloor=1e-9)
    timeVector[2,count] = @belapsed heppcat(Ylist_train,k,$iter; varfloor=1e-9)

    # LR-ALPCAH
    U_ALPCAH_LR = LR_ALPCAH(Ymat_train, k; alpcahIter=niters)
    timeVector[3,count] = @belapsed LR_ALPCAH(Ymat_train, k; alpcahIter=$iter)

    # QR-ALPCAH
    U_ALPCAH_QR = QR_ALPCAH(Ymat_train, k; alpcahIter=niters)
    timeVector[4,count] = @belapsed QR_ALPCAH(Ymat_train, k; alpcahIter=$iter)

    # ALPCAH
    U_ALPCAH = ALPCAH(Ymat_train, k, 10; μ=10, ρ=1.01, alpcahIter=niters)
    timeVector[5,count] = @belapsed ALPCAH(Ymat_train, k, 10; μ=10, ρ=1.01, alpcahIter=$iter)
    # subspace errors
    subspaces = [U_PCA, Mheppcat.U, U_ALPCAH_LR, U_ALPCAH_QR, U_ALPCAH]
    for (i,U) in enumerate(subspaces)
        test_errors[i,count] = norm(Ymat_test - U*(U'*Ymat_test)) / norm(Ymat_test)
    end
    count = count + 1
end
```

```julia
plot(title="Environmental Air Data - Time Comparisons",xlabel="Iterations", ylabel="Time (s)", legendfontsize=15, guidefontsize=15, titlefontsize=15)
groupedbar!(["1","10","50"],timeVector', bar_position = :dodge, bar_width=0.7, label = ["PCA" "HEPPCAT" "LR-ALPCAH" "QR-ALPCAH" "ALPCAH"])
#savefig("results_time.pdf")
```

```julia
# All methods converge very quickly so not including accuracy comparisons in paper
plot(title="Environmental Air Data - Accuracy Comparisons",xlabel="Iterations", ylabel="Error", legendfontsize=15, guidefontsize=15, titlefontsize=15)
groupedbar!(["1","10","50"],test_errors', bar_position = :dodge, bar_width=0.7, label = ["PCA" "HEPPCAT" "LR-ALPCAH" "QR-ALPCAH" "ALPCAH"])
#savefig("results_accuracy.pdf")
```
