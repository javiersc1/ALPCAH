```julia
using LinearAlgebra
using HePPCAT
using Plots
using ProgressMeter

using Random, Statistics, StatsPlots
using Measures

using PGFPlotsX
pgfplotsx()
#pyplot()
Plots.scalefontsizes()
Plots.scalefontsizes(0.75)
```

```julia
k = 5

niters = 100;
splitSize = round(Int, 50)
num_trials = 50;

num_methods = 4
test_errors = zeros(num_methods,num_trials)
train_errors = zeros(num_methods,num_trials)

p = Progress(num_trials; dt=1)
#Threads.@threads for iter=1:num_trials
for iter=1:num_trials
    trainShuffle = shuffle(1:size(Y)[2])[1:splitSize]
    testShuffle = shuffle(1:size(Y)[2])[1:splitSize]

    Ymat_train = Y[:,trainShuffle]
    Ymat_test = Y[:,testShuffle]

    ### Homoscedastic PPCA both groups
    U_PCA = PCA(Ymat_train, k)

    #HEPPCAT UNGROUPED
    Ylist_train = []
    for t=1:splitSize
        push!(Ylist_train, Ymat_train[:,t])
    end
    Mheppcat = heppcat(Ylist_train,k,niters; varfloor=1e-9)

    # LR-ALPCAH
    U_ALPCAH_LR = LR_ALPCAH(Ymat_train, k; alpcahIter=niters)

    # ALPCAH
    U_ALPCAH = ALPCAH(Ymat_train, k, 10; μ=10, ρ=1.01, alpcahIter=5)

    # errors = []
    subspaces = [U_PCA, Mheppcat.U, U_ALPCAH_LR, U_ALPCAH]
    for (i,U) in enumerate(subspaces)
        test_errors[i,iter] = norm(Ymat_test - U*(U'*Ymat_test)) / norm(Ymat_test)
        train_errors[i,iter] = norm(Ymat_train - U*(U'*Ymat_train)) / norm(Ymat_train)
    end
    next!(p)
end
finish!(p)
```

```julia
pgfplotsx()
boxplots = []
titles = ["Train Data", "Test Data"]
ylabels = ["NRMSE", "NRMSE"]
errors = [train_errors, test_errors];
for j=1:2
    p = plot(size=(500,300))
    p = boxplot(["PCA" "HePPCAT" "LR-ALPCAH" "ALPCAH"], errors[j]', leg = false,title=titles[j],color=[:green :purple :blue :orange],
        ylabel=ylabels[j], yticks = :native)
#     p = ylims!((0,0.4))
    push!(boxplots,p)  
end

plot_boxes = plot(boxplots...,layout=(1,2),size=(1800,400),left_margin = 0mm,
                    xtickfont=font(10), 
                    ytickfont=font(10), 
                    guidefont=font(10), 
                    legendfont=font(10))
#savefig("train_results.svg")
```
