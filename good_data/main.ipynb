```julia
using LinearAlgebra
using Plots
using Random
using Distributions
using Polynomials
using BenchmarkTools
using LaTeXStrings
using ProgressMeter
using JLD
```

```julia
include("../helperFunctions.jl")
include("../alpcahFamily.jl")
```

```julia
D = 50 # ambient space dimension
d = 5 # subspace dimension
goodPoints = [50;75;100;150;250;350;500;650;750;1000]#floor.(Int,range(50,1000, resolution))# good points per subspace
resolution = length(goodPoints)#100
ν1 = 1 # noise variance good points
ν2 = 20 # variance ratio for heatmaps
badPoints = 50
trials = 25; # smoother results
```

```julia
seedVector1 = rand(Xoshiro(1234), 1:100000, trials*resolution);
seedVector2 = rand(Xoshiro(4321), 1:100000, trials*resolution);
```

```julia
alpcahMatrix = zeros(resolution,trials)
pcaMatrix = zeros(resolution,trials)
pcaGoodMatrix = zeros(resolution,trials)

count = 1
@showprogress 1 "Running..." for i = 1:resolution # varianceRatio
    for k = 1:trials
        # subspace generation
        U = generateSubspace(D, d; seed=seedVector1[count]) 
        # data generation
        Y = generateData(U, [ν1,ν2], [goodPoints[i],badPoints]; seed=seedVector2[count])
        # pca
        U_PCA = PCA(Y,d)
        # PCA GOOD
        U_PCA_GOOD = PCA(Y[:,1:goodPoints[i]], d)
        # alpcah
        U_ALPCAH = LR_ALPCAH(Y,d; alpcahIter= 100)
        # error calculations
        pcaMatrix[i,k] = affinityError(U, U_PCA)
        alpcahMatrix[i,k] = affinityError(U, U_ALPCAH)
        pcaGoodMatrix[i,k] = affinityError(U, U_PCA_GOOD)
        # increase count
        count = count + 1
    end
end

save("trialData.jld", "alpcah", alpcahMatrix, "pca", pcaMatrix);
```

```julia
alpcahMean = vec(mean(alpcahMatrix, dims=2))
pcaMean = vec(mean(pcaMatrix,dims = 2))
pcaGoodMean = vec(mean(pcaGoodMatrix,dims = 2))


alpcahStd = vec(std(alpcahMatrix, dims=2))
pcaStd = vec(std(pcaMatrix, dims=2));
pcaGoodStd = vec(std(pcaGoodMatrix, dims=2));
```

```julia
color = palette(:default)
```

```julia
using Plots.Measures
#alpcahMean[1] = pcaMean[1]
#alpcahStd[1] = pcaStd[1]
plot(xlabel="N1, N2=$badPoints", ylabel=L"\Vert U_{*} U_{*}^{'}- UU' \Vert_F/ \Vert UU' \Vert_F", legendfontpointsize=12,xtickfontsize=12,ytickfontsize=10,guidefontsize=16,titlefontsize=18, legend=false)
plot!(goodPoints,pcaMean,ribbon=pcaStd, label="PCA", linewidth=3, color=color[3], fillalpha=0.3)
plot!(goodPoints,alpcahMean, ribbon=alpcahStd,label="LR-ALPCAH", linewidth=3, color=color[end-2], fillalpha=0.3)
#plot!(goodPoints,pcaGoodMean, ribbon=pcaGoodStd,label="PCAGOOD", linewidth=3, color=:black, fillalpha=0.3)
```
