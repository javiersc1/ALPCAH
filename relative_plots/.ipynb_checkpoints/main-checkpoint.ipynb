# Libraries

```julia
using Plots
using Random
using Distributions
using LinearAlgebra
using ProgressMeter
using JLD
```

```julia
include("../alpcahFamily.jl")
include("../helperFunctions.jl")
```

# Settings

```julia
varianceRatio = range( 1, 300,20) #20 for plot
pointRatio = range(1,100,20) # 20 points for plot

goodPoints = 20
ν1 = 0.1
dimSubspace = 5
ambientSpace = 100
trials = 50
```

# HEATMAP GENERATION (PCA and PCA-GOOD)

```julia
alpcahMatrix = zeros(trials,length(varianceRatio), length(pointRatio))
pcaMatrix = zeros(trials,length(varianceRatio), length(pointRatio))
pcaGoodMatrix = zeros(trials,length(varianceRatio), length(pointRatio))

@showprogress for i = 1:length(varianceRatio) # varianceRatio
    for j = 1:length(pointRatio) # pointRatio
        Threads.@threads for k = 1:trials
            ν2 = varianceRatio[i]*ν1
            badPoints = round(Int, pointRatio[j]*goodPoints)
            U = generateSubspace(ambientSpace, dimSubspace)
            Y = generateData(U, [ν1, ν2], [goodPoints, badPoints]; coordinateWindow = 10, coordinateType = :uniform) 
            # PCA
            U_PCA = PCA(Y, dimSubspace)
            pcaMatrix[k,i,j] = affinityError(U, U_PCA)
            
            # PCA GOOD POINTS ONLY
            U_good = 0
            if ν2 == ν1
                U_good = PCA(Y, dimSubspace)
            else
                U_good = PCA(Y[:,1:goodPoints], dimSubspace)
            end
            pcaGoodMatrix[k,i,j] = affinityError(U, U_good)
            
            # # ALPCAH
            # U_ALPCAH = LR_ALPCAH(Y, dimSubspace; alpcahIter=100)
            # alpcahMatrix[k,i,j] = affinityError(U, U_ALPCAH)
            
            # # ALPCAH KNOWN GROUPS
            # U_ALPCAH = LR_ALPCAH_GROUPED(Y,dimSubspace; alpcahIter = 100, goodpts=goodPoints)
            # alpcahMatrix[k,i,j] = affinityError(U, U_ALPCAH)

            # ALPCAH KNOWN VARIANCE
            w = vcat(ν1*ones(goodPoints), ν2*ones(badPoints))
            U_ALPCAH = LR_ALPCAH_KNOWN(Y,dimSubspace, w; alpcahIter = 100)
            alpcahMatrix[k,i,j] = affinityError(U, U_ALPCAH)
        end
    end
end
```

```julia
save("main_heatmaps_goodpts20.jld", "alpcahMatrix", alpcahMatrix, "pcaMatrix", pcaMatrix, "pcaGoodMatrix", pcaGoodMatrix)
```

```julia
xtickVector = [1;25;50;75;100]
ytickVector = [1;100;200;300];
```

```julia
alpcahMean = mean(alpcahMatrix, dims=1)[1,:,:]
pcaMean = mean(pcaMatrix, dims=1)[1,:,:]
pcaGoodMean = mean(pcaGoodMatrix, dims=1)[1,:,:];
```

```julia
data = alpcahMean./pcaMean
heatmap(pointRatio, varianceRatio, data, title="", xlabel="Point Ratio N2/N1, N1=$goodPoints", ylabel="Variance Ratio ν2/ν1, ν1=$ν1",yflip=true, clim=(minimum(data), 1.0), color=:viridis, xticks=xtickVector, yticks=ytickVector,
xtickfontsize=12,ytickfontsize=12,guidefontsize=12,titlefontsize=12)
#savefig("alpcah_known_pca_20goodpts.pdf")
```

```julia
data = alpcahMean./pcaGoodMean
heatmap(pointRatio, varianceRatio, data, title="", xlabel="Point Ratio N2/N1, N1=$goodPoints", ylabel="Variance Ratio ν2/ν1, ν1=$ν1",yflip=true, clim=(minimum(data),1.0), xticks=xtickVector, yticks=ytickVector,
xtickfontsize=12,ytickfontsize=12,guidefontsize=12,titlefontsize=12)
#savefig("alpcah_known_pca_good_20goodpts.pdf")
```

# GOOD DATA EXPERIMENT

```julia
varianceRatio = range( 1, 300,20) #20 for plot
goodPoints = floor.(Int,range(50,1000,20)) # 20 points for plot

badPoints = 50
ν1 = 0.1
dimSubspace = 5
ambientSpace = 100
trials = 50
```

```julia
alpcahMatrix = zeros(trials,length(varianceRatio), length(goodPoints))
pcaMatrix = zeros(trials,length(varianceRatio), length(goodPoints))

@showprogress for i = 1:length(varianceRatio) # varianceRatio
    for j = 1:length(goodPoints) # pointRatio
        Threads.@threads for k = 1:trials
            ν2 = varianceRatio[i]*ν1
            U = generateSubspace(ambientSpace, dimSubspace)
            Y = generateData(U, [ν1, ν2], [goodPoints[j], badPoints]; coordinateWindow = 10, coordinateType = :uniform) 
            # PCA
            U_PCA = PCA(Y, dimSubspace)
            pcaMatrix[k,i,j] = affinityError(U, U_PCA)

            # ALPCAH KNOWN VARIANCE
            w = vcat(ν1*ones(goodPoints[j]), ν2*ones(badPoints))
            U_ALPCAH = LR_ALPCAH_KNOWN(Y,dimSubspace, w; alpcahIter = 100)
            alpcahMatrix[k,i,j] = affinityError(U, U_ALPCAH)
        end
    end
end
```

```julia
save("good_data_goodpts20.jld", "alpcahMatrix", alpcahMatrix, "pcaMatrix", pcaMatrix)
```

```julia
xtickVector = [50;250;500;750;1000]
ytickVector = [1;100;200;300];
```

```julia
alpcahMean = mean(alpcahMatrix, dims=1)[1,:,:]
pcaMean = mean(pcaMatrix, dims=1)[1,:,:];
```

```julia
data = alpcahMean./pcaMean
heatmap(goodPoints, varianceRatio, data, title="", xlabel="N1, N2=$badPoints", ylabel="Variance Ratio ν2/ν1, ν1=$ν1",yflip=true, clim=(0.0, 1.0), xticks=xtickVector, yticks=ytickVector,
xtickfontsize=12,ytickfontsize=12,guidefontsize=12,titlefontsize=12)
#savefig("relative_good_data.pdf")
```

```julia
heatmap(goodPoints, varianceRatio, pcaMean - alpcahMean , title="", xlabel="N1, N2=$badPoints", ylabel="Variance Ratio ν2/ν1, ν1=$ν1",yflip=true, clim=(0.0, 1.0), xticks=xtickVector, yticks=ytickVector,
xtickfontsize=12,ytickfontsize=12,guidefontsize=12,titlefontsize=12)
#savefig("absolute_good_data.pdf")
```
